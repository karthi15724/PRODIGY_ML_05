```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
```

```python
model = MobileNetV2(weights="imagenet")
print("Model loaded successfully")
```

```python
# Sample food classes with average calories (per serving)
food_calories = {
    "pizza": 266,
    "burger": 295,
    "sandwich": 250,
    "pasta": 220,
    "salad": 150,
    "fried_rice": 300,
    "ice_cream": 207,
    "cake": 350,
    "donut": 260,
    "sushi": 200
}
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("food.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (224, 224))
img_array = img / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Predict food
prediction = model.predict(img_array)
predicted_index = np.argmax(prediction)

food_labels = list(food_calories.keys())
predicted_food = food_labels[predicted_index % len(food_labels)]
estimated_calories = food_calories[predicted_food]

# Display result
plt.imshow(img)
plt.axis("off")
plt.title(f"Food: {predicted_food}\nEstimated Calories: {estimated_calories} kcal")
plt.show()
```
